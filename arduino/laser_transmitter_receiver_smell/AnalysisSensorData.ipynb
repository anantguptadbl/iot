```python
import pandas as pd
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = []
for cur_file in glob.glob("./*.txt"):
    cur_data = pd.read_csv(cur_file, sep='->')
    cur_data['category'] = cur_file.replace("./", "").replace(".txt", "")
    cur_data.columns=['time', 'val', 'category']
    data.append(cur_data)
    
data=pd.concat(data, axis=0)

for cur_category in data['category'].unique():
    data[data['category']==cur_category]['val'].hist()
    plt.title("Category : {0}".format(cur_category))
    plt.show()
```
